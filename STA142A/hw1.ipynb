```python
list_beta=[]
i=0

random.seed(2)
while i < 1000:
    e = norm.rvs(0, 1, size=100)
    x = np.random.uniform(0,1,100)
    y = 5+3*x+e
    beta = (x.dot(y) - y.mean() * x.sum() ) / ( x.dot(x) - x.mean() * x.sum())
    list_beta.append(beta)
    i += 1

print(list_beta)
```

```python
beta_mean=sum(list_beta)/len(list_beta)
print('The mean of beta values is :',beta_mean)
```

```python
plt.hist(list_beta, 45) #histogram
```
